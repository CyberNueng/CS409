```python
import math
import numpy as np
```

```python
def logistic(x):
  ans=[]
  l=1
  k=1
  x0=1
  for i in xrange(len(x)):
    ans.append(1/(l+math.pow(math.e,(-k*(x[i]-x0)))))
  return ans
```

```python
x=np.arange(-10, 10, 0.1)
y=np.tanh(x)
z=logistic(x)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('RED=tanh, BLUE=logistic')
```
